### create a model to run through radlite

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from radmc3dPy import *
%matplotlib inline
```

```python
# set up the model and write out new style files for radmc
# note you need to have dustkappa_silicate.inp in this directory
analyze.writeDefaultParfile('ppdisk')
setup.problemSetupDust(model='ppdisk', mdisk='1e-2*ms', gap_rin='[0.1*au]', gap_rout='[0.5*au]', gap_drfact='[1e-5]', nz='0', binary=False)
```

```python
os.system("radmc3d mctherm")
```

```python
# translate temperature array to old radmc format
Tdust = data.dusttemp[:,:,0,0]
nr, nt = Tdust.shape
fname = "dusttemp_final.dat"
with open(fname, "w") as wfile:
    print("Writing " + fname)
    wfile.write(f"   1    {nr:d}   {nt//2:d}   1\n")
    wfile.write(" \n")
    wfile.write("   1\n")
    for i in range(nr):
        for j in range(nt//2):
            wfile.write(f"{Tdust[i,j]:.7f}\n")
os.system("mv "+fname+" radmc_outputs/")
            
fname = "dusttemp.info"
with open(fname, "w") as wfile:
    print("Writing " + fname)
    wfile.write("  -2\n")
    wfile.write("   1\n")
    wfile.write("   1\n")
    wfile.write("   1\n")
    wfile.write("   1\n")
os.system("mv "+fname+" radmc_outputs/")
```

```python
# write out the old format radmc files since radlite needs these :(
setup.problemSetupDust(model='ppdisk', mdisk='1e-2*ms', gap_rin='[0.1*au]', gap_rout='[0.5*au]', gap_drfact='[1e-5]', nz='0', binary=False, old=True)
```

```python
filelist = ["problem_params.inp", "dustopac.inp", "frequency.inp", "dustopac_1.inp", "frequency.inp", "radius.inp", "theta.inp", "starinfo.inp", "starspectrum.inp"]
for file in filelist:
    os.system("mv "+file+" radmc_outputs/")
```

```python
data = analyze.readData(dtemp=True)
```

```python
c1 = plt.contourf(data.grid.x/natconst.au, np.pi/2.-data.grid.y, data.dusttemp[:,:,0,0].T, 30, cmap='jet')
plt.xlabel('r [AU]')
plt.ylabel(r'$\pi/2-\theta$')
plt.xscale('log')
cb = plt.colorbar(c1)
cb.set_label('T [K]', rotation=270.)

c2 = plt.contour(data.grid.x/natconst.au, np.pi/2.-data.grid.y, data.dusttemp[:,:,0,0].T, 10,  colors='k', linestyles='solid')
plt.clabel(c2, inline=1, fontsize=10)
```
